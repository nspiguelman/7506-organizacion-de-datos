```python
import pandas as pd
import numpy as np
import random as rd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

%matplotlib inline
plt.style.use('default')
plt.rcParams['figure.figsize'] = (20, 10)
sns.set(style="whitegrid")
pd.options.display.float_format = '{:20,.2f}'.format
warnings.filterwarnings('ignore')
```

```python
train_labels = pd.read_csv('../../datasets/train_labels.csv')
```

```python
train_values = pd.read_csv('../../datasets/train_values.csv')
```

```python
escuelas = train_values.loc[train_values['has_secondary_use_school']==1,'building_id']
```

```python
grado_danio_escuelas = train_labels.merge(escuelas, how='inner')
```

```python
grado_danio_escuelas = grado_danio_escuelas\
    .groupby('damage_grade')\
    .agg('count')\
    .rename(columns={'building_id':'cantidad_de_escuelas'})
```

```python
grado_danio_escuelas
```

```python
grado_danio_escuelas.plot(
    kind='bar',
    figsize=(10,10),
    rot=0,
    title='Número de escuelas con cada grado de daño',
    color=(0.2, 0.4, 0.6, 0.6)
)
ax=plt.gca()
ax.set_xlabel('damage_grade')
ax.set_ylabel('Cantidad de escuelas');
```
